# Matplotlib example workflow

```python
#0 import matplotlib and get it ready for plotting
%matplotlib inline
import matplotlib.pyplot as plt

#1 Prepare Data
x = [1, 2, 3, 4]
y = [10, 20, 30, 40]

#2 Setup Plot
fig, ax = plt.subplots(figsize=(10, 10)) # 10, 10 is x(width), y(width)

#3 Plot Data
ax.plot(x, y)

#4 Customize Plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

#5 Save & show (save the whole figure)
fig.savefig("./plots/sample_plot.png")
```

```python
# Make a plot from dictionary
butter_prices = {'Almond Butter': 10,
                 'Cashew butter': 12,
                 'Peanut Butter': 8}
fig, ax = plt.subplots()
ax.bar(butter_prices.keys(), butter_prices.values())
ax.set(title="Butter Store",
       ylabel="Prices ($)");
```

```python
# Horizontal Bar
fig, ax = plt.subplots()
ax.barh(list(butter_prices.keys()), list(butter_prices.values()))
```

```python
# Histogram
x = np.random.randn(1000) # normal distribution
fig, ax = plt.subplots()
ax.hist(x);
```

### Two Options for Subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))

# Plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(butter_prices.keys(), butter_prices.values());
ax4.hist(np.random.randn(100));
```

```python
# Subplots option 2
fig , ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10, 5))
# Plot
ax[0,0].plot(x , x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(butter_prices.keys(), butter_prices.values());
#ax[1,1].barh(list(butter_prices.keys()), list(butter_prices.values()));
ax[1,1].hist(np.random.randn(1000));
```

### pyplot vs matplotlib OO method?
* when plotting something quickly, okay to use pyplot method
* when plotting something more advanced, Use OO method.

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
len(over_50)
```

```python
# Pyplot Method
over_50.plot(kind="scatter",
             x='age',
             y='chol',
             c='target',
             figsize=(10,6));
```

```python
# OO Method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter",
             x='age',
             y='chol',
             c='target',
             ax=ax);
#ax.set_xlim([45, 100]);
```

```python
over_50["target"].values
```

```python
## OO Method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# Plot The Data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"]);

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");

# Add a Legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle='--');
```

```python
over_50.head()
```

```python
 # Subplot of chol, age and thalach
fig , (ax0, ax1) = plt.subplots(nrows=2,
                                ncols=1,
                                figsize=(10,10),
                                sharex=True)

#Add Data to  ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"]);

ax0.set(title="Heart Disease and Cholesterol",
        ylabel="Cholesterol")
ax0.legend(*scatter.legend_elements(), title="Target")
ax0.axhline(y=over_50["chol"].mean(),
            linestyle='--');

# Add Data to ax1
scatter2 = ax1.scatter(x=over_50["age"],
                       y=over_50["thalach"],
                       c=over_50["target"])
ax1.set(title="Heart Disease and Max Heart Rate",
        xlabel="Age",
        ylabel="Max Heart Rate")
ax1.legend(*scatter2.legend_elements(), title="Target");
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--");

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
plt.style.use("seaborn-v0_8-whitegrid")

## OO Method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# Plot The Data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter"); # this changes the color schemem

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");

# Add a Legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle='--');
```

```python
# Customizing the x and y axes limitations
 # Subplot of chol, age and thalach
fig , (ax0, ax1) = plt.subplots(nrows=2,
                                ncols=1,
                                figsize=(10,10),
                                sharex=True)

#Add Data to  ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter");

ax0.set(title="Heart Disease and Cholesterol",
        ylabel="Cholesterol")
ax0.legend(*scatter.legend_elements(), title="Target")
ax0.axhline(y=over_50["chol"].mean(),
            linestyle='--');
ax0.set_xlim([50, 80])
# Add Data to ax1
scatter2 = ax1.scatter(x=over_50["age"],
                       y=over_50["thalach"],
                       c=over_50["target"],
                       cmap="winter")
ax1.set(title="Heart Disease and Max Heart Rate",
        xlabel="Age",
        ylabel="Max Heart Rate")
ax1.legend(*scatter2.legend_elements(), title="Target");
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--");
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])
# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
