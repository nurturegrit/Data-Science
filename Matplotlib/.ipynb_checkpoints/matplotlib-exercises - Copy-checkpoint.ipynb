```python
# Understanding Exponential Plot, Let us start with a smaller range for x
import numpy as np
import matplotlib.pyplot as plt

# Define a practical range for x
x = np.arange(0, 10, 1)  # x values from 0 to 9

# Calculate the exponential values
y = np.exp(x)

# Create the plot
fig, ax = plt.subplots()
ax.plot(x, y, label="exp(x)")

# Adding title and labels
ax.set(title="Plot of Exponential Numbers",
       xlabel="Integers",
       ylabel="Exponentials")

# Adding a legend
ax.legend()

# Display the plot
plt.show()

# Print the exponential values for reference
print(y)
```

The y-axis displaying values like [0.0, 0.5, 1.0, ... 2.5] with a label of 1e43 at the end is a common way of handling very large numbers in plots. The 1e43 notation is scientific notation for 1×10431×1043, which indicates that the values on the y-axis are being scaled by this factor.

This can be confusing if you are not used to seeing plots with such large numbers. Let's go through the steps to properly understand and visualize the exponential growth over a large range.
Using a Logarithmic Scale for the y-axis

To make the plot easier to interpret, you can use a logarithmic scale for the y-axis. This will compress the range of the y-values, making the exponential growth more manageable to visualize.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range for x
x = np.arange(1, 101, 1)  # x values from 1 to 100

# Calculate the exponential values
y = np.exp(x)

# Create the plot
fig, ax = plt.subplots()
ax.plot(x, y, label="exp(x)")

# Adding title and labels
ax.set(title="Plot of Exponential Numbers",
       xlabel="Integers",
       ylabel="Exponentials")

# Set y-axis to logarithmic scale
ax.set_yscale('log')

# Adding a legend
ax.legend()

# Display the plot
plt.show()

# Print the exponential values for reference
print(y)
```

Explanation:

    Range Selection:
        x = np.arange(1, 101, 1): This creates an array of integers from 1 to 100. Starting from 1 avoids the exponential of 0, which is 1.

    Calculate Exponential Values:
        y = np.exp(x): Computes the exponential values for the given x array.

    Plot Creation:
        fig, ax = plt.subplots(): Initializes the plot.
        ax.plot(x, y, label="exp(x)"): Plots the exponential function.

    Adding Titles and Labels:
        ax.set(title="Plot of Exponential Numbers", xlabel="Integers", ylabel="Exponentials"): Adds a title and labels for the x and y axes.

    Logarithmic Scale:
        ax.set_yscale('log'): Sets the y-axis to a logarithmic scale, making the large range of exponential values easier to visualize.

    Display and Print:
        plt.show(): Displays the plot.
        print(y): Prints the calculated exponential values for reference.

By using a logarithmic scale for the y-axis, the exponential growth is displayed in a way that is easier to interpret, especially over a large range of x values. This approach helps in visualizing the exponential function without the overwhelming scale differences.

```python
# Import the '../data/car-sales.csv' into a DataFame called car_sales and view
car_sales = pd.read_csv("../Pandas/car-sales.csv")
car_sales
```

```python
# Try to plot the 'Price' column using the plot() function
car_sales["Price"].plot()
```

Why doesn't it work?

Hint: It's not numeric data.

In the process of turning it to numeric data, let's create another column which adds the total amount of sales and another one which shows what date the car was sold.

Hint: To add a column up cumulatively, look up the cumsum() function. And to create a column of dates, look up the date_range() function.

```python
# Remove the symbols, the final two numbers from the 'Price' column and convert it to numbers
car_sales["Price"] = car_sales["Price"].astype(str).str.replace('[\$,]', '', regex=True).astype(float).astype(int)
car_sales
```

```python
# Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column
car_sales["Total Sales"] = car_sales["Price"].cumsum()

# Add a column called 'Sale Date' which lists a series of successive dates starting from today (your today)
car_sales["Sale Date"] = pd.date_range('02-08-2024', periods=len(car_sales))
# View the car_sales DataFrame
car_sales
```

Now we've got a numeric column (`Total Sales`) and a dates column (`Sale Date`), let's visualize them.

```python
# Use the plot() function to plot the 'Sale Date' column versus the 'Total Sales' column
car_sales.plot('Sale Date', 'Total Sales');
```

```python
# Convert the 'Price' column to the integers


# Create a scatter plot of the 'Odometer (KM)' and 'Price' column using the plot() function
car_sales.plot.scatter("Price", 'Odometer (KM)');
```

Now let's try comparing two variables versus the target variable.

More specifially we'll see how age and cholesterol combined effect the target in **patients over 50 years old**.

For this next challenge, we're going to be replicating the following plot:

<img src="../images/matplotlib-heart-disease-chol-age-plot.png"/>

```python
# Replicate the above plot in whichever way you see fit
# Note: The method below is only one way of doing it, yours might be
# slightly different

# Create DataFrame with patients over 50 years old
over_50 = heart_disease[heart_disease['age'] > 50]

# Create the plot
fig, ax = plt.subplots()

# Plot the data
scatter = ax.scatter(x=over_50['age'],
           y=over_50['chol'],
           c=over_50['target'],
           cmap='winter');

# Customize the plot
ax.set(title="Effects of Cholesterol on Heart Diseases (Over Age 50)",
       xlabel="Age",
       ylabel="Cholesterol")
ax.legend(*scatter.legend_elements(), title="Target")
# Add a meanline
ax.axhline(y=over_50['chol'].mean(), label="Mean Cholesterol")
```

Beatiful, now you've created a plot of two different variables, let's change the style.

```python
# Check what styles are available under plt
plt.style.available
```

```python
# Change the style to use "seaborn-whitegrid"
plt.style.use('seaborn-v0_8-whitegrid')
```

Now the style has been changed, we'll replot the same figure from above and see what it looks like.

If you've changed the style correctly, it should look like the following:
<img src="../images/matplotlib-heart-disease-chol-age-plot-seaborn-whitegrid.png"/>


```python
# Reproduce the same figure as above with the "seaborn-whitegrid" style
fig, ax = plt.subplots()
# Create the plot
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap='winter');

# Customize the plot
ax.set(title="Effects of Cholesterol on Heart Diseases for People over 50 age",
       xlabel="Age",
       ylabel="Cholesterol");
ax.legend(*scatter.legend_elements(), title="Targets")
# Add a meanline
ax.axhline(y=over_50['chol'].mean());
```

Wonderful, you've changed the style of the plots and the figure is looking different but the dots aren't a very good colour.

Let's change the `cmap` parameter of `scatter()` as well as the `color` parameter of `axhline()` to fix it.

Completing this step correctly should result in a figure which looks like this:
<img src="../images/matplotlib-heart-disease-chol-age-plot-cmap-change.png"/>

```python
# Replot the same figure as above except change the "cmap" parameter
fig, ax = plt.subplots()
# Create the plot
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap='winter');

# Customize the plot
ax.set(title="Effects of Cholesterol on Heart Diseases for People over 50 age",
       xlabel="Age",
       ylabel="Cholesterol");
ax.legend(*scatter.legend_elements(), title="Targets")
# Add a meanline
ax.axhline(y=over_50['chol'].mean(), color='red');
```

Beautiful! Now our figure has an upgraded color scheme let's save it to file.

```python
# Save the current figure using savefig(), the file name can be anything you want
fig.savefig('plots/matplotlib_exercise.png')
```
